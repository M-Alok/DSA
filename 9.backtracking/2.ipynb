### Generate Parentheses

Link - [Leetcode-22](https://leetcode.com/problems/generate-parentheses)

```python
from typing import List

def generateParenthesis(n: int) -> List[str]:
    res = []
    stack = []

    def dfs(opened: int, closed: int):
        if opened == closed == n:
            res.append(''.join(stack))
            return

        if opened < n:
            stack.append('(')
            dfs(opened+1, closed)
            stack.pop()
        
        if closed < opened:
            stack.append(')')
            dfs(opened, closed+1)
            stack.pop()
    dfs(0, 0)
    return res

generateParenthesis(3)
```

### Word Search

Link - [Leetcode-79](https://leetcode.com/problems/word-search)

```python
def exist(board: List[List[str]], word: str) -> bool:
    ROWS, COLS = len(board), len(board[0])

    def dfs(r: int, c: int, i: str, path: set):
        if i == len(word):
            return True
        
        if (r < 0 or r >= ROWS or c < 0 or c >= COLS or
            (r, c) in path or word[i] != board[r][c]):
            return False
        
        path.add((r, c))
        res = (dfs(r+1, c, i+1, path) or 
               dfs(r-1,c, i+1, path) or 
               dfs(r, c+1, i+1, path) or 
               dfs(r, c-1, i+1, path))
        path.remove((r, c))
        return res
    
    for r in range(ROWS):
        for c in range(COLS):
            if dfs(r, c, 0, set()):
                return True
    return False

board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
word = "ESEC"
exist(board, word)
```

### Palindrome Partitioning

Link - [Leetcode-131](https://leetcode.com/problems/palindrome-partitioning)

```python
from typing import List

def isPal(s: str, l: int, r: int) -> bool:
    while l < r:
        if s[l] != s[r]:
            return False
        l, r = l+1, r-1
    return True

def partition(s: str) -> List[List[str]]:
    res = []

    def dfs(i: int, path: list):
        if i == len(s):
            res.append(path[:])
            return
        
        for j in range(i, len(s)):
            if isPal(s, i, j):
                path.append(s[i:j+1])
                dfs(j+1, path)
                path.pop()
    dfs(0, [])
    return res

partition('aab')
```

### Letter Combinations of a Phone Number

Link - [Leetcode-17](https://leetcode.com/problems/letter-combinations-of-a-phone-number/description/)

```python
def letterCombinations(digits: str) -> List[str]:
    letters = {
        '2': 'abc', '3': 'def', '4': 'ghi',
        '5': 'jkl', '6': 'mno', '7': 'pqrs',
        '8': 'tuv', '9': 'wxyz',
    }
    res = []

    def dfs(i: int, cur: str):
        if len(cur) == len(digits):
            res.append(cur)
            return
        
        for ch in letters[digits[i]]:
            dfs(i+1, cur + ch)
    if digits:
        dfs(0, '')
    return res

letterCombinations('23')
```

### N-Queens

Link - [Leetcode-51](https://leetcode.com/problems/n-queens)

```python
from typing import List

def solveNQueens(n: int) -> List[List[str]]:
    cols = set()
    posDiag = set()
    negDiag = set()

    res = []
    board = [['.'] * n for _ in range(n)]

    def dfs(r):
        if r == n:
            copy = [''.join(row) for row in board]
            res.append(copy)
            return
        
        for c in range(n):
            if c in cols or (r+c) in posDiag or (r-c) in negDiag:
                continue

            cols.add(c)
            posDiag.add(r+c)
            negDiag.add(r-c)
            board[r][c] = 'Q'
            
            dfs(r+1)
            
            cols.remove(c)
            posDiag.remove(r+c)
            negDiag.remove(r-c)
            board[r][c] = '.'
    dfs(0)
    return res

solveNQueens(4)
```
